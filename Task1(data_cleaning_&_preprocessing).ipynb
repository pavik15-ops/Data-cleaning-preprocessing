# **handle missing values**

```python
df.isnull().sum()
```

# **Remove duplicate**

```python
df.drop_duplicates(inplace=True)
print(df)
```

# **Rename column headers**

```python
df.rename(columns={
    'ID': 'Customer_ID',
    'Year_Birth': 'Birth_Year',
    'Education': 'Education_Level',
    'Marital_Status': 'Marital_Status',
    'Income': 'Annual_Income',
    'Kidhome': 'Kids_At_Home',
    'Teenhome': 'Teens_At_Home',
    'Dt_Customer': 'Customer_Since',
    'Recency': 'Days_Since_Last_Purchase',
    'MntWines': 'Spent_On_Wines',
    'NumWebVisitsMonth': 'Monthly_Web_Visits',
    'AcceptedCmp3': 'Accepted_Campaign_3',
    'AcceptedCmp4': 'Accepted_Campaign_4',
    'AcceptedCmp5': 'Accepted_Campaign_5',
    'AcceptedCmp1': 'Accepted_Campaign_1',
    'AcceptedCmp2': 'Accepted_Campaign_2'
}, inplace=True)
print(df)
```

# **Fix data types**

```python
df["ID"]=df["ID"].astype(int)
```

# **Summary of Changes :**

*   Removed Duplicates(There are no duplicates in the given dataset)
*   Fixed Data types for the Id column

*   Checked whether there are missing values or not
*   Renamed all the column names into proper standard form for better understanding
